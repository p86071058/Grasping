```python
ddtf = '/content/drive/My Drive/project/Clean data/TF/TF' + str(14).zfill(2)
list_dir = ('drink1')
data_length = []
fn = ddtf + '/' + list_dir
fdf = pd.read_csv(fn, sep = '\t',skiprows = range(26))
```

```python
fdf['T_resultant'] = (fdf.T_fx**2 + fdf.T_fy**2 + fdf.T_fz**2)**(1/2)
fdf['I_resultant'] = (fdf.I_fx**2 + fdf.I_fy**2 + fdf.I_fz**2)**(1/2)
fdf['M_resultant'] = (fdf.M_fx**2 + fdf.M_fy**2 + fdf.M_fz**2)**(1/2)
fdf['R_resultant'] = (fdf.R_fx**2 + fdf.R_fy**2 + fdf.R_fz**2)**(1/2)
fdf['L_resultant'] = (fdf.L_fx**2 + fdf.L_fy**2 + fdf.L_fz**2)**(1/2)
fdf['T_contribute']= (fdf.T_resultant)/(fdf.T_resultant+fdf.I_resultant+fdf.M_resultant+fdf.R_resultant+fdf.L_resultant)
fdf['I_contribute']= (fdf.I_resultant)/(fdf.T_resultant+fdf.I_resultant+fdf.M_resultant+fdf.R_resultant+fdf.L_resultant)
fdf['M_contribute']= (fdf.M_resultant)/(fdf.T_resultant+fdf.I_resultant+fdf.M_resultant+fdf.R_resultant+fdf.L_resultant)
fdf['R_contribute']= (fdf.R_resultant)/(fdf.T_resultant+fdf.I_resultant+fdf.M_resultant+fdf.R_resultant+fdf.L_resultant)
fdf['L_contribute']= (fdf.L_resultant)/(fdf.T_resultant+fdf.I_resultant+fdf.M_resultant+fdf.R_resultant+fdf.L_resultant)
```

fz fx fy --> fr ft fl

```python
list_chanel = ['T_fx','T_fy','T_fz','I_fx','I_fy','I_fz','M_fx','M_fy','M_fz','R_fx','R_fy','R_fz','L_fx','L_fy','L_fz','ACC_x','ACC_y','ACC_z']
list_chanel_resultant = ['T_resultant','I_resultant','M_resultant','R_resultant','L_resultant']
list_chanel_contribute = ['T_contribute','I_contribute','M_contribute','R_contribute','L_contribute']
fdf = fdf[list_chanel + list_chanel_resultant + list_chanel_contribute]
fdf
```

```python
## Calculate features
df_mean = fdf[list_chanel + list_chanel_resultant + list_chanel_contribute].mean()
df_std = fdf[list_chanel + list_chanel_resultant + list_chanel_contribute].std()
df_min = fdf[list_chanel + list_chanel_resultant + list_chanel_contribute].min()
df_max = fdf[list_chanel + list_chanel_resultant+list_chanel_contribute].max()
df_median = fdf[list_chanel + list_chanel_resultant + list_chanel_contribute].median()
df_mode = fdf[list_chanel + list_chanel_resultant + list_chanel_contribute].mode()
df_mode = df_mode.iloc[0]
df_skew = fdf[list_chanel + list_chanel_resultant + list_chanel_contribute].skew()
df_kurt = fdf[list_chanel + list_chanel_resultant + list_chanel_contribute].kurt()
```

```python
count_row = fdf.shape[0]  # gives number of row count
count_col = fdf.shape[1]  # gives number of col count
fdf.shape
```

```python
data_sumary = pd.concat([df_mean,df_std,df_min,df_max,df_median,df_mode,df_skew,df_kurt],keys=['mean','std','min','max','median','mode','Skewness','Kurtosis'],axis=1)
data_sumary['Range'] = data_sumary['max']-data_sumary['min']
print('Data length',fdf.shape[0])
numb_para = data_sumary.shape[0]*data_sumary.shape[1] +1
print('Number of parameter: ',numb_para)
print(data_sumary.shape)
data_sumary
```
